```python
import numpy as np
import pandas as pd
```

```python
movie_titles = pd.read_csv(r"movies.csv")
movie_titles.head()
```

```python
movie_ratings = pd.read_csv(r"ratings.csv")
movie_ratings.head()
```

```python
movie_titles.drop("genres", inplace = True, axis = 1)
movie_titles.head()
```

```python
movie_ratings.drop("timestamp", inplace = True, axis = 1)
movie_ratings.head()
```

```python
merged_movie_df = pd.merge(movie_ratings, movie_titles, on='movieId')
merged_movie_df.head()
```

```python
user_movie_rating_matrix = merged_movie_df.pivot_table(index='userId', columns='title', values='rating')
user_movie_rating_matrix
```

```python
interaction = user_movie_rating_matrix.T
interaction = interaction.fillna(0)

interaction = interaction[interaction.sum(axis=1)>0]
interaction
```

```python
def dot_product(vector1,vector2):
 thedotproduct=np.sum([vector1[k]*vector2[k] for k in range(len(vector1))])
 return(thedotproduct)
```

```python
def vector_norm(vector):
 thenorm=np.sqrt(dot_product(vector,vector))
 return(thenorm)
```

```python
def cosine_similarity(vector1,vector2):
 thedotproduct=dot_product(vector1,vector2)
 thecosine=thedotproduct/(vector_norm(vector1)*vector_norm(vector2))
 thecosine=np.round(thecosine,4)
 return(thecosine)
```

```python
def get_item_recommendations(interaction,itemname, n=5):
    otherrows=[rowname for rowname in interaction.index if rowname!=itemname]
    otheritems=interaction.loc[otherrows,:]
    theitem=list(interaction.loc[itemname,:])
    similarities=[]
    for items in otheritems.index:
        similarities.append(cosine_similarity(theitem,list(otheritems.loc[items,:])))
    otheritems['similarities']=similarities
    recoms=list(otheritems.sort_values(by='similarities',ascending=False).index)
    return recoms[:n]
```

```python
print("LIST OF MOVIES::")
for m in list(list(interaction.index)):
  print(m)
```

```python
movie_name = input("Enter the movie name:")
num_of_recs=int(input("Enter the number of recommendations:"))
recommendations= get_item_recommendations(interaction,movie_name,num_of_recs)

print("Your top "+str(num_of_recs)+" recommendations for "+movie_name+" are:")
count=1
for m in recommendations:
    print(str(count)+". "+m)
    count+=1
```

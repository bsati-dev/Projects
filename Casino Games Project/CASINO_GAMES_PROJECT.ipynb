##**ROULETTE GAME:**

```python
import random
import matplotlib.pyplot as plt
import numpy as np
import math
def roulette(rounds = 100, bet_amt=5) :
    net_balance = 0
    earnings = bet_amt * 35
    #simulate each round for i in range(rounds):
        #bet straight up and get winning number
        #37 = 00
    bet_num = random.randint(0,37)
    win_num = random.randint(0,37)
    #adjust balance
    if bet_num == win_num:
        net_balance += earnings
    else:
        net_balance -= bet_amt
    return net_balance
```

```python
#simulate 100, 1000, and 10000 rounds
datal = np.array([roulette() for i in range (10000)])

#adjust bin width on histogram
bin_width = 40
min_val = datal.min()
max_val = datal.max()
bins = np.arange(min_val, max_val + bin_width, bin_width)

#create histogram
plt.hist(datal, bins=bins)
plt.xlabel("Net balance after 1000 rounds")
plt.ylabel ("Simulation count")
plt.title("Roulette wealth distribution")
# plt.show()
```

```python
import math

n = 100         # number of trials
p = 1/38        # probability of winning
q = 1 - p       # probability of losing

exp_x = 0       # expected value
exp_xsqr = 0    # expected value of X^2

for k in range(n + 1):
    # Probability of k wins in n trials
    prob = math.comb(n, k) * (p ** k) * (q ** (n - k))

    # Value function: net gain/loss
    value = (k * 180) - 500

    # Update expected value and expected square
    exp_x += prob * value
    exp_xsqr += prob * (value ** 2)

# Variance
variance = exp_xsqr - (exp_x ** 2)

print(f"Expected value: {exp_x:.2f}")
print(f"Variance: {variance:.2f}")
```

##**CRAPS GAME:**

```python
import random

def roll_dice():
    return random.randint(1,6) + random.randint(1,6)

def playing_craps(bet):
    print("ðŸŽ²STARTING CRAPSðŸŽ²")

    #FIRST ROLL
    roll = roll_dice()
    print("First Roll:", roll)

    if roll in [7,11]:
        print("PASS LINE WINSðŸ‘‘")
        return bet

    elif roll in [2,3,12]:
        print("PASS LINE LOSESðŸ¤¡")
        return -bet
    else:
        point = roll
        print("The Point has been set to:",point)

    #SECOND ROLL

    while True:
        roll = roll_dice()
        print("Rolled:",roll)

        if roll == point:
            print("Shooter hit the POINT --> WINSðŸ‘‘")
            return bet

        elif roll == 7:
            print("Shooter hit a 7 --> LOSESðŸ¤¡")
            return -bet

        else:
            print("ðŸ˜µROLL AGAINðŸ˜µ")

def craps_rounds(bet,rounds,start_money):
    money=start_money
    money_track=[money]

    for x in range(rounds):
        res= playing_craps(bet)
        money+=res
        money_track.append(money)

    return money_track


#EXAMPLE::

start_money=10000
bet = 100
rounds = 100

money_track = craps_rounds(bet,rounds,start_money)

print("Starting wealth:", start_money)
print("Wealth after playing",rounds,"rounds",money_track[-1])
print("Profit will be:",money_track[-1]-start_money)

```
